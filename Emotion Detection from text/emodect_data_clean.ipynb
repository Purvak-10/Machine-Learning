## Text Cleaning
1. Remove emojis
2. Converting to Lowercase
3. Removing Text in Square Brackets
4. Removing Links
5. Removing Punctuation
6. Removing Words Containing Numbers
7. Removing Special Characters
8. Expand Contractions
9. Correcting Spelling Errors
10. Removing Extra Spaces

```python
# spelling checker
spell = SpellChecker(language='en')

# To clean text
def clean_text(text):
    print(text)
    text = remove_emojis(text)  # 1
    text = text.lower()  # 2
    text = re.sub(r'\[.*?\]', '', text)  # 3
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # 4
    text = re.sub(r'[^\w\s]', '', text)  # 5
    text = re.sub(r'\w*\d\w*', '', text)  # 6
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # 7
    text = contractions.fix(text)  # 8
    text = ' '.join([spell.correction(word) if spell.correction(word) else word for word in text.split()])  # 9
    print(text,"/o/")

    text = re.sub(r'\s+', ' ', text).strip()  # 10

    return text

# to remove emojis from the text
def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags
        "\U00002702-\U000027B0"  # miscellaneous symbols
        "\U000024C2-\U0001F251"  # enclosed characters
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
# Call the clean_text() to the "text" column
df['cleaned_text'] = df['text'].apply(clean_text)
```
